```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=db.drop(['Lat','Long'],axis=1,inplace=True)
```

```python
corona_aggre=db.groupby("Country/Region").sum()
```

```python
corona_aggre.head()
```

```python
corona_aggre.shape
```

```python
corona_aggre.loc['China'].plot()
corona_aggre.loc['India'].plot()
corona_aggre.loc['Spain'].plot()
plt.legend()
```

```python
corona_aggre.loc['China'].diff().plot()
```

```python
corona_aggre.loc['China'].diff().max()
```

```python
countries = list(corona_aggre.index)
max_infection_rates=[]
for c in countries:
    max_infection_rates.append(corona_aggre.loc[c].diff().max())
corona_aggre['Max_infection_rates']=max_infection_rates
```

```python
corona_aggre
```

```python
corona_data=pd.DataFrame(corona_aggre["Max_infection_rates"])
```

```python
corona_data
```

```python
happiness_report=pd.read_csv("/Users/rishi/Downloads/worldwide_happiness_report.csv")
```
